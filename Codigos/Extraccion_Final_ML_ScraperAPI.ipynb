### FASE 1 – Obtener lista de productos desde URLs de paginación

```python
# Lista vacía para guardar los productos encontrados en cada página
productos = []

# Recorrer cada página de resultados
for i, url in enumerate(urls_paginadas, 1):
    print(f"Procesando página {i} de {len(urls_paginadas)}")

    # Hacer la solicitud a través de ScraperAPI
    response = requests.get(SCRAPER_API_URL.format(url))

    # Verificar que la respuesta sea válida
    if response.status_code != 200:
        print(f"Error en URL: {url}")
        continue

    # Analizar el contenido HTML de la respuesta
    soup = BeautifulSoup(response.text, 'html.parser')

    # Buscar los contenedores de productos
    items = soup.find_all('li', {'class': 'ui-search-layout__item'})

    # Extraer información básica de cada producto
    for item in items:
        try:
            descripcion = item.find('a', {'class': 'poly-component__title'}).text.strip()
            url_producto = item.find('a', href=True)['href'].split('#')[0]  # Limpiar URL de parámetros
            productos.append({
                'descripcion': descripcion,
                'url_producto': url_producto
            })
        except Exception as e:
            print(f"Error extrayendo producto: {e}")

    # Esperar entre 3 y 7 segundos para evitar bloqueo
    time.sleep(random.uniform(3, 7))

# Convertir la lista de productos en un DataFrame
df = pd.DataFrame(productos)
```

### FASE 2 – Extraer info detallada del producto

```python
# Función para extraer el ID del producto desde la URL
def extraer_id(url):
    match = re.search(r"MCO-\d+", url)
    return match.group(0).replace("-", "") if match else None

# Crear una columna nueva con el ID del producto
df['id_producto'] = df['url_producto'].apply(extraer_id)

# Crear columna de URL para preguntas usando el ID del producto
df['url_preguntas'] = df['id_producto'].apply(
    lambda x: f"https://articulo.mercadolibre.com.co/noindex/questions/{x}?scroll_to_question=2&new_version=true&modal=false" if x else None
)

```

```python
df.sample(5)  # Mostrar las primeras filas del DataFrame con la nueva columna
```

```python
# Lista para almacenar la información de cada producto
detalles = []

# Iterar sobre cada URL de producto
for i, url in enumerate(df['url_producto'], 1):
    print(f"Procesando producto {i}/{len(df)}")

    # Hacer solicitud a través de ScraperAPI
    try:
        response = requests.get(SCRAPER_API_URL.format(url))
        response.raise_for_status()
    except Exception as e:
        print(f"Error en la solicitud de {url}: {e}")
        detalles.append({
            'descripcion': "No disponible",
            'vendidos': "No disponible",
            'precio': "No disponible",
            'calificacion': "No disponible",
            'cant_calificaciones': "No disponible",
            'valoracion': "No disponible",
            'vendor': "No disponible",
            'cant_comentarios': "No disponible",
            'comentarios': [],
            'vendor_info': "No disponible",
        })
        continue

    # Parsear el HTML con BeautifulSoup
    soup = BeautifulSoup(response.text, 'html.parser')

    # Buscar los distintos elementos en la página
    nombre = soup.find('h1', {'class': 'ui-pdp-title'})
    vendidos = soup.find('div', {'class': 'ui-pdp-header__subtitle'})
    precio = soup.find('span', {'class': 'andes-money-amount__fraction'})
    calificacion = soup.find('p', {'class': 'andes-visually-hidden'})
    cant_calificaciones = soup.find('p', {'class': 'ui-review-capability__rating__label'})
    valoracion = soup.find('p', {'class': 'ui-review-capability__rating__average ui-review-capability__rating__average--desktop'})
    vendor_elem = soup.find('span', {'class': 'ui-pdp-seller__label-sold'})
    vendor_inf = soup.find('span', {'class': 'ui-pdp-color--BLACK ui-pdp-size--LARGE ui-pdp-family--REGULAR ui-seller-data-header__title non-selectable'})
    cant_comentarios = soup.find('span', {'class': 'total-opinion'})
    comentarios = soup.find_all('p', {'class': 'ui-review-capability-comments__comment__content ui-review-capability-comments__comment__content'})

    # Buscar nombre del vendedor si es posible
    if vendor_elem:
        vendedor_nombre_elem = vendor_elem.find_next('span')
        vendor_nombre = vendedor_nombre_elem.text.strip() if vendedor_nombre_elem else "No disponible"
    else:
        vendor_nombre = "No disponible"

    # Guardar los datos extraídos en la lista
    detalles.append({
        'descripcion': nombre.text.strip() if nombre else "No disponible",
        'vendidos': vendidos.text.strip() if vendidos else "No disponible",
        'precio': precio.text.strip() if precio else "No disponible",
        'calificacion': calificacion.text.strip() if calificacion else "No disponible",
        'cant_calificaciones': cant_calificaciones.text.strip() if cant_calificaciones else "No disponible",
        'valoracion': valoracion.text.strip() if valoracion else "No disponible",
        'vendor': vendor_nombre,
        'cant_comentarios': cant_comentarios.text.strip() if cant_comentarios else "No disponible",
        'comentarios': [c.text.strip() for c in comentarios] if comentarios else [],
        'vendor_info': vendor_inf.text.strip() if vendor_inf else "No disponible",
    })

    # Esperar entre 3 y 5 segundos para evitar bloqueos
    time.sleep(random.uniform(3, 5))

# Convertir a DataFrame y unir al principal
df_detalles = pd.DataFrame(detalles)
df = pd.concat([df, df_detalles], axis=1)
```

### FASE 3 – Extracción de preguntas y respuestas

```python
# Inicializar columnas vacías
df["preguntas"] = None
df["respuestas"] = None

# Recorrer cada URL de preguntas del DataFrame
for i, (idx, url) in enumerate(df['url_preguntas'].items(), 1):
    print(f"Extrayendo preguntas {i}/{len(df)}")

    # Verificar que la URL sea válida
    if not isinstance(url, str) or not url.startswith("http"):
        continue

    try:
        # Hacer la solicitud vía ScraperAPI
        response = requests.get(SCRAPER_API_URL.format(url), timeout=20)

        if response.status_code != 200:
            continue

        # Analizar el contenido HTML de la página de preguntas
        soup = BeautifulSoup(response.text, 'html.parser')

        # Extraer todas las preguntas
        preguntas = [p.text.strip() for p in soup.find_all('span', {
            'class': 'ui-pdp-color--BLACK ui-pdp-size--SMALL ui-pdp-family--REGULAR ui-pdp-questions__questions-list__question'})]

        # Extraer todas las respuestas
        respuestas = [r.text.strip() for r in soup.find_all('span', {
            'class': 'ui-pdp-color--GRAY ui-pdp-size--SMALL ui-pdp-family--REGULAR ui-pdp-questions__questions-list__answer-item__separate ui-pdp-questions__questions-list__answer-item__with-tapeables-links'})]

        # Asignar al DataFrame según el índice original
        df.at[idx, 'preguntas'] = preguntas if preguntas else []
        df.at[idx, 'respuestas'] = respuestas if respuestas else []

    except Exception as e:
        print(f"Error en {url}: {e}")

    # Pausa aleatoria entre solicitudes
    time.sleep(random.uniform(3, 7))
```
