```python
# Test extraccion listado de Urls productos, url de preguntas y respuestas a partir de la url de producto (Expreciones regulares)



# Define headers, para simulación de un navegador y evitar bloqueos.
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36'
}

# Lista vacía donde vamos a guardar los datos.
data = []

# Ciclo para recorrer las URLs
for i, url in enumerate(urls, 1):
    print(f"Processing {i} of {len(urls)}")

    try:
        # Enviar una solicitud HTTP para obtener el contenido de la página.
        response = requests.get(url, headers=headers)

        # Si la respuesta no es exitosa, imprimir mensaje y continuar con la siguiente URL
        if response.status_code != 200:
            print(f"Error {response.status_code} en {url}. Saltando...")
            continue  # Pasa a la siguiente URL sin detener la ejecución
        
        # Parsear el contenido HTML de la página usando BeautifulSoup
        soup = BeautifulSoup(response.text, 'html.parser')

        # Extraer los contenedores de los productos.
        productos = soup.find_all('li', {'class': 'ui-search-layout__item'})

        # Ciclo para recorrer cada producto y extraer la información.
        for producto in productos:
            try:
                # Extraer el nombre del producto.
                descripcion = producto.find('a', {'class': 'poly-component__title'}).text.strip()

                # Extraer URL del producto.
                url_producto = producto.find('a', href=True)['href']

                # Agregar la información a la lista de datos.
                data.append({
                    'descripcion': descripcion,
                    'url_producto': url_producto,
                })
            except AttributeError:
                print("Error al extraer datos del producto. Saltando...")  # Mensaje de error, pero sigue con los demás productos
                continue

    except requests.exceptions.RequestException as e:
        print(f"Error en la solicitud HTTP: {e}. Saltando...")  # Captura errores de conexión, tiempo de espera, etc.
        continue  # Sigue con la siguiente URL
```

### Construir el identificador de id de producto y pagina de preguntas y respuestas a partir de expresiones regulares.

```python
# Codigo para extraer id de producto a partir de la url y crear la url de preguntas y respuestas.
import re

# Definir una función para extraer el ID del producto a partir de la URL
def extraer_id(url):
    match = re.search(r"MCO-\d+", url)
    return match.group(0).replace("-", "") if match else None

# Aplicar la función a la columna 'url_producto' para extraer el ID del producto
df['id_producto'] = df['url_producto'].apply(extraer_id)

# Crear la URL de preguntas y respuestas correctamente usando apply()
df['url_preguntas'] = df['id_producto'].apply(lambda x: f"https://articulo.mercadolibre.com.co/noindex/questions/{x}?scroll_to_question=2&new_version=true&modal=false" if x else None)

df.head()
```

```python
import requests
from bs4 import BeautifulSoup
import time

# Headers para simular un navegador y evitar bloqueos
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36'
}

# Lista vacía para almacenar los datos extraídos
data_productos = []

# Recorrer las URLs del DataFrame
for i, url in enumerate(df['url_producto'], 1):
    print(f"Processing {i} of {len(df['url_producto'])}")
    response = requests.get(url, headers=headers)
    
    if response.status_code != 200:
        print(f"Skipping URL {url} due to status code {response.status_code}")
        continue  # Saltar la iteración si la respuesta no es exitosa
    
    soup = BeautifulSoup(response.text, 'html.parser')
    
    # Extraer el nombre del producto
    descripcion_elem = soup.find('h1', {'class': 'ui-pdp-title'})
    descripcion = descripcion_elem.text.strip() if descripcion_elem else "No disponible"

    # Extraer cantidad de vendidos del producto
    vendidos_elem = soup.find('div', {'class' :'ui-pdp-header__subtitle'})
    vendidos = vendidos_elem.text.strip() if vendidos_elem else "No disponible"
    
    # Extraer el precio del producto
    precio_elem = soup.find('span', {'class': 'andes-money-amount__fraction'})
    precio = precio_elem.text.strip() if precio_elem else "No disponible"
    
    # Extraer la calificación completa del producto
    calificacion_elem = soup.find('p', {'class': 'andes-visually-hidden'})
    calificacion_completa = calificacion_elem.text.strip() if calificacion_elem else "No disponible"
    
    # Extraer la cantidad de calificaciones del producto
    calificaciones_elem = soup.find('p', {'class': 'ui-review-capability__rating__label'})
    calificaciones = calificaciones_elem.text.strip() if calificaciones_elem else "No disponible"
    
    # Extraer la valoración del producto
    valoracion_elem = soup.find('p', {'class': 'ui-review-capability__rating__average ui-review-capability__rating__average--desktop'})
    valoracion = valoracion_elem.text.strip() if valoracion_elem else "No disponible"

    # Extraer calificación desde opiniones del superior
    calificacion_superior_elem = soup.find('span', {'class': 'ui-pdp-review__rating'})
    calificacion_superior = calificacion_superior_elem.text.strip() if calificacion_superior_elem else "No disponible"

    # Extraer cantidad de calificaciones del superior
    calificaciones_superior_elem = soup.find('span', {'class': 'ui-pdp-review__amount'})
    calificaciones_superior = calificaciones_superior_elem.text.strip() if calificaciones_superior_elem else "No disponible"

    # Extraer nombre del vendedor del superior
    vendor_superior_elem = soup.find('span', {'class': 'ui-pdp-seller__label-sold'})
    if vendor_superior_elem:
        vendedor_nombre_elem = vendor_superior_elem.find_next('span')  # Buscar el siguiente <span>
        vendor_superior = vendedor_nombre_elem.text.strip() if vendedor_nombre_elem else "No disponible"
    else:
        vendor_superior = "No disponible"

    # Extraer nombre del vendedor del inferior
    vendor_inferior_elem = soup.find('span', {'class': 'ui-pdp-color--BLACK ui-pdp-size--LARGE ui-pdp-family--REGULAR ui-seller-data-header__title non-selectable'})
    vendor_inferior = vendor_inferior_elem.text.strip() if vendor_inferior_elem else "No disponible"

    # Extraer cantidad de comentarios del producto.
    #comentario_cantidad_elem = soup.find('span', {'class': 'ui-pdp-review__amount'})
    comentario_cantidad_elem = soup.find('span',{'class' : 'total-opinion'})
    comentario_cantidad = comentario_cantidad_elem.text.strip() if comentario_cantidad_elem else "No disponible"

    # Extraer todos los comentarios disponibles.
    comentarios_elem = soup.find_all('p', {'class': 'ui-review-capability-comments__comment__content ui-review-capability-comments__comment__content'})
    comentarios = [comentario.text.strip() for comentario in comentarios_elem] if comentarios_elem else ["No disponible"]
    
    # Agregar la información a la lista
    data_productos.append({
        'descripcion': descripcion,
        'vendidos': vendidos,
        'precio': precio,
        'calificacion_completa': calificacion_completa,
        'calificaciones': calificaciones,
        'valoracion': valoracion,
        'calificacion_superior': calificacion_superior,
        'calificaciones_superior': calificaciones_superior,
        'vendor_superior': vendor_superior,
        'vendor_inferior': vendor_inferior,
        'comentario_cantidad': comentario_cantidad,
        'comentarios': comentarios
    })

    # Pausa para evitar bloqueos
    time.sleep(3)

print("Scraping completado.")
```

## Fase 3. Extraccion de preguntas y respuestas. 

```python
# Crear una lista vacía para almacenar las preguntas y respuestas
data_preguntas = []

# Definir headers para evitar bloqueos en las solicitudes
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36'
}

# Recorrer las URLs de preguntas y respuestas
for i, url in enumerate(df['url_preguntas'], 1):
    print(f"Processing {i} of {len(df['url_preguntas'])}")
    
    # Verificar si la URL es válida
    if not isinstance(url, str) or not url.startswith("http"):
        print(f"Skipping invalid URL: {url}")
        data_preguntas.append({'preguntas': [], 'respuestas': []})  # Agregar lista vacía para mantener el orden
        continue  # Saltar esta iteración si la URL no es válida

    # Enviar una solicitud HTTP para obtener el contenido de la página con headers
    response = requests.get(url, headers=headers)
    
    if response.status_code != 200:
        print(f"Skipping URL {url} due to status code {response.status_code}")
        data_preguntas.append({'preguntas': [], 'respuestas': []})  # Agregar lista vacía para mantener el orden
        continue  # Saltar esta iteración si la solicitud no fue exitosa
    
    soup = BeautifulSoup(response.text, 'html.parser')
    
    # Extraer las preguntas
    preguntas = [p.text.strip() for p in soup.find_all('span', {'class': 'ui-pdp-color--BLACK ui-pdp-size--SMALL ui-pdp-family--REGULAR ui-pdp-questions__questions-list__question'})]
    
    # Extraer las respuestas
    respuestas = [r.text.strip() for r in soup.find_all('span', {'class': 'ui-pdp-color--GRAY ui-pdp-size--SMALL ui-pdp-family--REGULAR ui-pdp-questions__questions-list__answer-item__separate ui-pdp-questions__questions-list__answer-item__with-tapeables-links'})]

    # Agregar la información alineada con el DataFrame original
    data_preguntas.append({'preguntas': preguntas, 'respuestas': respuestas})

    # Pausa para evitar bloqueos
    time.sleep(2)

print("Scraping completado.")
```

## Fase 4 Exploracion de los datos recolectados.

```python
import ast

# Convertir los valores de 'comentarios', 'preguntas' y 'respuestas' en listas reales si están en formato de string
def safe_eval(x):
    if isinstance(x, str):
        try:
            return ast.literal_eval(x)
        except:
            return []
    return x if isinstance(x, list) else []

df['comentarios'] = df['comentarios'].apply(safe_eval)
df['preguntas'] = df['preguntas'].apply(safe_eval)
df['respuestas'] = df['respuestas'].apply(safe_eval)

# Contar la cantidad de elementos en cada lista, asegurándonos de que sean listas antes de aplicar len()
df['cantidad_comentarios'] = df['comentarios'].apply(lambda x: len(x) if isinstance(x, list) else 0)
df['cantidad_preguntas'] = df['preguntas'].apply(lambda x: len(x) if isinstance(x, list) else 0)
df['cantidad_respuestas'] = df['respuestas'].apply(lambda x: len(x) if isinstance(x, list) else 0)

df.head()
```
